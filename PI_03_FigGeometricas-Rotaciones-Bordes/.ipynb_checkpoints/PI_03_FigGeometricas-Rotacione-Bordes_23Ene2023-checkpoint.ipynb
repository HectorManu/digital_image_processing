# 0. Librerías básicas

```python
#Librerías básicas de esta notebook
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv
```

# 1. Figuras geométricas


### Creación de una imagen en color negro

```python
# Imagen color negro de 512 x 512 
canvas = np.zeros(shape=(512,512,3),dtype=np.int16)
canvas.shape
```

```python
plt.imshow(canvas)
plt.show()
```

#### Rectangulo

```python
#cv.rectangle(imagen,origen,final,color,grosor de linea)
#cv.rectangle(canvas,pt1=(100.10),pt2=(200,110),color = (0,255,0),thickness =5)
cv.rectangle( canvas , pt1=(100,10) , pt2=(200,110) , 
             color=( 0,255,0  ) ,
             thickness=1)
plt.imshow(canvas)
plt.show()
```

```python
cv.rectangle( canvas , pt1=(255,255) , pt2=(500,500) , 
             color=( 0,0,255 ) ,
             thickness=4)
plt.imshow(canvas)
plt.show()
```

#### Circulo

```python
#cv.circle(imagen, centro, radio, color, grosor de linea)
    # Imagen color negro de 512 x 512 
canvas = np.zeros(shape=(512,512,3),dtype=np.int16)
canvas.shape


cv.circle(canvas, center=(250,250) , radius = 10 , color = (0,255,0), thickness = 3)

cv.circle(canvas, center=(250,250) , radius = 10 , color = (255,0,0), thickness = 1)

cv.circle(canvas, center=(250,250) , radius = 100 , color = (255,0,0), thickness = 3)

plt.imshow(canvas)
plt.show()
```

#### Linea

```python
#cv.line(imagen, inicio, fin, color, grosor de linea)
# Imagen color negro de 512 x 512 
canvas = np.zeros(shape=(512,512,3),dtype=np.int16)
canvas.shape

cv.line(canvas, pt1=(0,0), pt2 = (250,250), color = (255,255,255), thickness = 3) 

cv.line(canvas, pt1=(512,0), pt2 = (250,250), color = (255,255,255), thickness = 3) 


plt.imshow(canvas)
plt.show()
```

#### Texto

```python
#font = cv.FONT_HERSHEY_SIMPLEX
#lineType=cv.LINE_AA
#cv.putText(imagen,texto,origen, fuente,tamaño de la fuente,color,grosor, tipo de linea)
# Imagen color negro de 512 x 512 
canvas = np.zeros(shape=(512,512,3),dtype=np.int16)
canvas.shape


cv.putText(canvas,text='Hola Mundo',
           org=(100, 200), 
           fontFace = cv.FONT_HERSHEY_SIMPLEX,
           fontScale=1,
           color=(0,255,0),
           thickness=3,
           lineType=cv.LINE_AA)


plt.imshow(canvas)
plt.show()
```


#### Polígono 

```python
#cv.polylines(imagen,puntos,Cerrada o abierta,color y grosor)
# Imagen color negro de 512 x 512 
canvas = np.zeros(shape=(512,512,3),dtype=np.int16)
canvas.shape

#cv.polylines(canvas, (np.array([[100,200],[400,200],[400,500]])), 
#            isClosed = True,
#            color = (0,255,0),
#            thickness = 3
#            ) 

puntos=np.array(  [ [100,200] , [400,200] , [400,500]  ]   )
cv.polylines(canvas,  [ puntos ]  ,
             isClosed=True,
             color=(0,255,0),
             thickness=3
            )

plt.imshow(canvas)
plt.show()
```

# 2. Rotación de imagenes

## 2.1 Función *flip*

```python
img = cv.imread('figuras/wave.png')
img = cv.cvtColor( img, cv.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
rotacion_1=cv.flip(img,1)
plt.imshow(rotacion_1)
plt.show()
```

```python
rotacion_0=cv.flip(img,0)
plt.imshow(rotacion_0)
plt.show()
```

```python
rotacion_m1=cv.flip(img,-1)
plt.imshow(rotacion_m1)
plt.show()
```
