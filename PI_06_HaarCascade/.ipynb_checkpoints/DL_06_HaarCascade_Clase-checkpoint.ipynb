# Cargar librerias

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv
```

```python
img = cv.imread('figuras/bbt.jpeg')
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.show()
```

```python
def plt_imshow(title, image):
    image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
    plt.imshow(image)
    plt.title(title)
    plt.grid(False)
    plt.show()
```
